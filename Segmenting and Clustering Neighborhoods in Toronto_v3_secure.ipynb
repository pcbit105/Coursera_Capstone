```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html_results = requests.get(url).text
soup = BeautifulSoup(html_results,'lxml')
print(soup.prettify())
```

```python
canada_table = soup.find('table',{'class':'wikitable sortable'})
canada_table
```

```python
canada_headers = canada_table.findAll('th')
header_list = []
for header in canada_headers:
    header_list.append(header.text.rstrip())
header_list
```

```python
canada_data = canada_table.findAll('td')
postal_code_list = []
borough_list = []
neighbourhood_list = []
for data in canada_data[0::3]:
    postal_code_list.append(data.text.rstrip())
for data in canada_data[1::3]:
    borough_list.append(data.text.rstrip())
for data in canada_data[2::3]:
    neighbourhood_list.append(data.text.rstrip())
```

```python
canada_df2 = pd.DataFrame(columns=header_list)
canada_df2['Postal Code'] = postal_code_list
canada_df2['Borough'] = borough_list
canada_df2['Neighbourhood'] = neighbourhood_list
canada_df2
```

Now that we have the dataframe, we will need to process the data to make it usable.

```python
canada_df2 = canada_df2[canada_df2['Borough'] != 'Not assigned'].reset_index(drop=True)
canada_df2
```

Let's produce a new dataframe with dummy values for venue categories, then add a Neighbourhood column, followed by grouping by Neighbourhoods by mean values for each venue category.

```python
toronto_venues_dummies = pd.get_dummies(toronto_venues['Venue Category'])
toronto_venues_dummies.head(12)
```

```python
toronto_venues_dummies.insert(loc=0, column='Neighbourhood', value=toronto_venues['Neighbourhood'])
toronto_venues_dummies.head(50)
```

```python
toronto_venues_grouped = toronto_venues_dummies.groupby('Neighbourhood').mean().reset_index()
toronto_venues_grouped
```

Let's print each neighbourhoods with their top 5 most frequent venue categories

```python
num_top_venues = 5

for hood in toronto_venues_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = toronto_venues_grouped[toronto_venues_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

To put the top 10 most common venue categories for each neighbourhood into a dataframe, let's first define a function to sort the most common categories, and then use this function to create the dataframe.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighbourhood'] = toronto_venues_grouped['Neighbourhood']

for ind in np.arange(toronto_venues_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_venues_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

Let's use KMeans to segment the neighbourhoods into 5 clusters

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_venues_grouped.drop('Neighbourhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Let's add the clusters to the neighborhoods_venues_sorted dataframe

```python
neighborhoods_venues_sorted2 = neighborhoods_venues_sorted
neighborhoods_venues_sorted2.insert(loc=0, column='Cluster', value=kmeans.labels_)
neighborhoods_venues_sorted2
```

Let's then merge the dataframes so we can see all the geographic info, along with the corresponding clusters and top 10 most common venues

```python
toronto_df_clusters = toronto_df.join(neighborhoods_venues_sorted2.set_index('Neighbourhood'), on=toronto_df_clusters['Neighbourhood'])
toronto_df_clusters
```

Let's now create a map of Toronto, with marker colorings based on the clusters they belond to

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)

ys = [i + x + (i*x)**2 for i in range(kclusters)]

colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to map, differentiated by clusters
for lat, lng, borough, neighborhood, cluster in zip(toronto_df_clusters['Latitude'], toronto_df_clusters['Longitude'], toronto_df_clusters['Borough'], toronto_df_clusters['Neighbourhood'], toronto_df_clusters['Cluster']):
    label = '{}, {}, {}'.format(postal, neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
